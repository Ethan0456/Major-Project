```python
import pandas as pd
import glob

df_ls = []
for file in glob.glob('*.csv'):
    df = pd.read_csv(file)
    df.columns = ['S.N.', 'District', 'Site Name', 'Latitude', 'Longitude', 'pH', 'EC',
       'TDS', 'TH', 'CO3', 'HCO3', 'Cl', 'S04', 'NO3', 'F', 'Alkalinity', 'Ca',
       'Mg', 'Na', 'K', 'SiO2', 'SAR']
    df_ls.append(df)

df = pd.concat(df_ls).reset_index(drop=True)
df
```

```python
df.to_csv('../../refined-data/tabula-Gujarat_State_Year_Book_2021-22.csv', index=False)
```
