```python
import pandas as pd

data = pd.read_csv('./super_data.csv')
results = pd.read_csv('./wqi_results.csv')

data = pd.concat([data, results], axis=1)
data
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

corr = data.drop(['DISTRICT','LOCATION'], axis=1).corr()
# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
plt.figure(figsize=(15, 15))

# Choose a color palette
palette = sns.color_palette("Blues", n_colors=20)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=palette, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5}, annot=True, fmt=".2f")

plt.title('Correlation Plot', fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```
