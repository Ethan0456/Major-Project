```python
import pandas as pd

data = pd.read_csv('./refined-data/eval_data.csv')
results = pd.read_csv('./refined-data/wqi_results.csv')

data = pd.concat([data, results], axis=1)
data
```

```python
data = data.replace({
    'Ahmedabad':'Ahmadabad',
    'Bharuch*':'Bharuch',

    'Vadodara*' : 'Vadodara',
    'Vadodra':'Vadodara',

    'Jundagadh':'Junagadh',

    'Chhota udepur': 'Dahod',
    'Chhota Udaipur': 'Dahod',

    'Aravalli': 'Sabar Kantha',
    'Arvalli':'Sabar Kantha',

    'Morvi': 'Rajkot',
    'Morbi': 'Rajkot',

    'Dangs-Ahwa':'The Dangs',
    'The dangs':'The Dangs',
    'Dang': 'The Dangs',

    'Kutchch':'Kachchh',

    'Porbanadar':'Porbandar',

    'Panchmahals':'Panch Mahals',
    'Panchmahal': 'Panch Mahals',

    'Dohad':'Dahod',
    'Sabarkantha':'Sabar Kantha',
    'Banaskantha': 'Banas Kantha',

    'Tapi': 'Surat',
    'Diu': 'Amreli',
    'Daman': 'Valsad',
    'Mata no madh': 'Kachchh',
    'Mehsana': 'Mahesana',
    'Mahisagar': 'Panch Mahals',
    'Gir Somnath': 'Junagadh',
    'Botad': 'Ahmadabad',
    'Devbhumi Dwarka': 'Jamnagar',
})
district_data = data.groupby(['DISTRICT'])['WAWQI'].mean().reset_index(name='WAWQI')
district_data
```

```python
import geopandas as gpd

data = gpd.read_file(
    "./GeoJSON/india_district.geojson"
)

print(type(data))
```

```python
data_gujarat = data[data['NAME_1'] == "Gujarat"]
data_gujarat
```

```python
districts_2 = data_gujarat['NAME_2']
districts_2
```

```python
# import the folium library
import folium

# initialize the map and store it in a m object
m = folium.Map(location=[22.2587, 71.1924], zoom_start=7)
m
```

```python
folium.Choropleth(
    geo_data=data_gujarat,
    name="choropleth",
    data=district_data,
    columns=["DISTRICT", "WAWQI"],
    key_on="feature.properties.NAME_2",  # Adjust this according to your GeoJSON structure
    fill_color="BuGn",  # Adjust the color scale
    fill_opacity=1, 
    line_opacity=1,
    legend_name="WAWQI"
).add_to(m)

m.show_in_browser()
```
