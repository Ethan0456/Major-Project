```python
import pandas as pd

data = pd.read_csv("./super_data.csv")
results = pd.read_csv("./wqi_results.csv")

data = pd.concat([data, results], axis=1)
data
```

```python
data = data.replace({
    'Ahmedabad':'Ahmadabad',
    'Bharuch*':'Bharuch',

    'Vadodara*' : 'Vadodara',
    'Vadodra':'Vadodara',

    'Jundagadh':'Junagadh',
    'Junagarh': 'Junagadh',

    'Chhota udepur': 'Dahod',
    'Chhota Udaipur': 'Dahod',

    'Aravalli': 'Sabar Kantha',
    'Arvalli':'Sabar Kantha',

    'Morvi': 'Rajkot',
    'Morbi': 'Rajkot',

    'Dangs-Ahwa':'The Dangs',
    'The dangs':'The Dangs',
    'Dang': 'The Dangs',

    'Kutchch':'Kachchh',
    'Kacchh': 'Kachchh',

    'Porbanadar':'Porbandar',

    'Panchmahals':'Panch Mahals',
    'Panchmahal': 'Panch Mahals',

    'Dohad':'Dahod',
    'Sabarkantha':'Sabar Kantha',
    'Banaskantha': 'Banas Kantha',

    'Tapi': 'Surat',
    'Diu': 'Amreli',
    'Daman': 'Valsad',
    'Mata no madh': 'Kachchh',
    'Mehsana': 'Mahesana',
    'Mahisagar': 'Panch Mahals',
    'Gir Somnath': 'Junagadh',
    'Botad': 'Ahmadabad',
    'Devbhumi Dwarka': 'Jamnagar',
    'Porbander': 'Porbandar',
})
district_data = data.groupby(['DISTRICT','YEAR'])['WAWQI'].mean().reset_index(name="WAWQI")
district_data
```

```python
district_data.sort_values(by=['WAWQI'], ascending=False)
```

```python
# Group the data by year
grouped = district_data.groupby('YEAR')

# Iterate over each group (year)
for year, group in grouped:
    # Find district with minimum WAWQI for the year
    min_district = group.loc[group['WAWQI'].idxmin(), 'DISTRICT']
    min_wawqi = group['WAWQI'].min()
    
    # Find district with maximum WAWQI for the year
    max_district = group.loc[group['WAWQI'].idxmax(), 'DISTRICT']
    max_wawqi = group['WAWQI'].max()
    
    print(f"For year {year}:")
    print(f"District with minimum WAWQI: {min_district} (WAWQI: {min_wawqi})")
    print(f"District with maximum WAWQI: {max_district} (WAWQI: {max_wawqi})")
    print()
```
